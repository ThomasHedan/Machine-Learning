### Second dataset:

```python
Journeys_inter = df_inter[['Departure station','Arrival station','Average travel time (min)']].copy()
Journeys_inter['Average travel time (min)'].isna().sum()
```

```python
Journeys_inter.dropna(subset=['Average travel time (min)'],inplace=True)
Journeys_inter['Average travel time (min)'].isna().sum()
```

```python
def f(a,b):
    return Journeys_inter[(Journeys_inter['Departure station']==a) & (Journeys_inter['Arrival station']==b) | (Journeys_inter['Departure station']==b) & (Journeys_inter['Arrival station']==a)]['Average travel time (min)'].mean()

Journeys_inter['Travel time']= Journeys_inter.apply(lambda x : f(x['Departure station'],x['Arrival station']),axis=1)
Journeys_inter.drop(['Average travel time (min)'],axis=1,inplace=True)
```

```python
Journeys= Journeys_inter.drop_duplicates()
Journeys.head()
```

```python
Journeys.to_csv('Journeys.csv')
```

```python
def f(x):
    y=0
    if('Less than 5'==x):
        y=-5
    else:
        y=int(x)
    return y

df2.NB_VALID=df2.NB_VALID.apply(f)
```
